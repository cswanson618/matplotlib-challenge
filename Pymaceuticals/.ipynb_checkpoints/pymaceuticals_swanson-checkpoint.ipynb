```python
import matplotlib.pyplot as plt
import matplotlib.figure
import pandas as pd
import numpy as np
import os
import warnings
warnings.filterwarnings('ignore')

master_mouse_path = os.path.join("data", "mouse_drug_data.csv")
master_clinical_path = os.path.join("data", "clinicaltrial_data.csv")

master_mouse_df = pd.read_csv(master_mouse_path)
master_clinical_df = pd.read_csv(master_clinical_path)

master_df= master_mouse_df.merge(master_clinical_df, right_on= "Mouse ID", left_on = "Mouse ID", how= "outer")
```

```python
master_df
```

## Tumor Response to Treatment

```python
tumor_df = pd.pivot_table(master_df, values = "Tumor Volume (mm3)", index = ["Timepoint"], columns = ["Drug"])
tumor_df
#Stack overflow was glorious for teaching about Pandas Pivot Tables. We really should have covered this in class given how
#powerful it is
```

```python
tumor_sem_df = master_df.groupby(["Drug", "Timepoint"]).sem().drop(columns = ["Metastatic Sites", "Mouse ID"])
tumor_sem_df_final = tumor_sem_df.pivot_table(values= "Tumor Volume (mm3)", index = ["Timepoint"], columns = ["Drug"])
tumor_sem_df_final
```

```python
x = [0, 5, 10, 15, 20, 25, 30, 35, 40, 45]
y1= tumor_df["Capomulin"]
y2 = tumor_df["Ceftamin"]
y3= tumor_df["Infubinol"]
y4 = tumor_df["Ketapril"]
y5 = tumor_df["Naftisol"]
y6 = tumor_df["Placebo"]
y7 = tumor_df["Propriva"]
y8 = tumor_df["Ramicane"]
y9 = tumor_df["Stelasyn"]
y10= tumor_df["Zoniferol"] 
plt.plot(x, y1, marker = "o", label = "Capomulin")
plt.plot(x, y2, marker = "o", label = "Ceftamin")
plt.plot(x, y3, marker = "o", label = "Infubinol")
plt.plot(x, y4, marker = "o", label = "Ketapril")
plt.plot(x, y5, marker = "o", label = "Naftisol")
plt.plot(x, y6, marker = "o", label = "Placebo")
plt.plot(x, y7, marker = "o", label = "Propriva")
plt.plot(x, y8, marker = "o", label = "Ramicane")
plt.plot(x, y9, marker = "o", label = "Stelasyn")
plt.plot(x, y10, marker = "o", label = "Zoniferol")
plt.errorbar(x, y1, yerr= tumor_sem_df_final["Capomulin"])
plt.errorbar(x, y2, yerr= tumor_sem_df_final["Ceftamin"])
plt.errorbar(x, y3, yerr= tumor_sem_df_final["Infubinol"])
plt.errorbar(x, y4, yerr= tumor_sem_df_final["Ketapril"])
plt.errorbar(x, y5, yerr= tumor_sem_df_final["Naftisol"])
plt.errorbar(x, y6, yerr= tumor_sem_df_final["Placebo"])
plt.errorbar(x, y7, yerr= tumor_sem_df_final["Propriva"])
plt.errorbar(x, y8, yerr= tumor_sem_df_final["Ramicane"])
plt.errorbar(x, y9, yerr= tumor_sem_df_final["Stelasyn"])
plt.errorbar(x, y10, yerr= tumor_sem_df_final["Zoniferol"])
plt.xlabel("Timepoint (Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.ylim([30,75])
plt.xlim ([0, 50])
plt.title ("Tumor Volume Over Time")
plt.legend()
plt.rcParams["figure.figsize"] = (15,15)
plt.show()

#This is the only way I could get this (and subsequent) code to work. I'm sure there was a more elegant way to do this with
#loops but I couldn't get it to work. I tried plotting with tumor_df.plot, which worked great, but then wouldn't let the
#error bars work.
```

## Survival Rates

```python
survival_df = master_df.groupby(["Drug", "Timepoint"]).count().drop(columns = ["Metastatic Sites", "Tumor Volume (mm3)"])
survival_df_final = survival_df.pivot_table(values= "Mouse ID", index = ["Timepoint"], columns = ["Drug"])
```

### Raw Numbers

```python
survival_df_final
```

### By Percentage

```python
pct_survival = round((((survival_df_final) / (survival_df_final.iloc[0]))*100), 2)
```

```python
pct_survival
```

```python
pct_survival.plot(marker = "o")
plt.xlabel("Timepoint (Days)")
plt.ylabel("Percent Survival")
plt.title ("Survival Over Time")
plt.ylim([0,110])
plt.rcParams["figure.figsize"] = (15,15)
plt.show()
```

## Summary Bar Graph

```python
pct_change = round((((tumor_df.iloc[9])/(tumor_df.iloc[0]))*100)-100,2)
```

```python
pct_change_df = pd.DataFrame(pct_change)
pct_change_df1 = pd.DataFrame(pct_change[pct_change < 0])
pct_change_df2 = pd.DataFrame(pct_change[pct_change > 0])
drug_list1 = pct_change_df1.index.tolist()
drug_list2 = pct_change_df2.index.tolist()
pct_list1 = pct_change_df1[0].tolist()
pct_list2 = pct_change_df2[0].tolist()
```

```python
be0 = plt.bar(drug_list1, pct_list1, color = "green")
ab0 = plt.bar(drug_list2, pct_list2, color = "red")
plt.xlabel("Drug")
plt.ylabel("Percent Change")
plt.title("Tumor Size Change (%) Over 45 Day Period")
plt.xticks(rotation = 90)
plt.grid(axis = "y")
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        plt.text(rect.get_x() + rect.get_width()/2., .00*height,
                '%d' % int(height),
                ha='center', va='bottom')
        #credit for the autolabel function goes to the matplotlib api example code found at: 
        #https://matplotlib.org/examples/api/barchart_demo.html
autolabel(be0)
autolabel(ab0)
plt.show()
```

# Observations


1. It goes without saying that Capomulin and Ramicane are the two drugs with the most success. The most mice survive on them, they're the only two that shrunk tumors, they had the fewest metastatic sites over time, and had had the lowest tumor volume over time

2. Something might be wrong, potentially toxic, with Propriva, Infubinol, and Zoniferol. Survival rates are strongly disproportionte with the tumor growth/volume compared to other drugs that showed continued tumor growth. Survival rates are severely worse over the 45 days, especially Propriva and Zoniferol, which show a significant, early survival rate drop-off. Infubinol's dropoff comes off later, perhaps suggesting toxicity from continued use.

3. Ketapril seems to make things <i>worse</i> for tumor growth and spread. Its tumor volume growth over time/percentage change are relatively above placebo and there's a marked increase in metastatic sites above placebo.
